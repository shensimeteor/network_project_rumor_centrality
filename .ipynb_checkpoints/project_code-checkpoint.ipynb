```python
import random
class SIModel2:
     # 
    def __init__(self, graph, init_infected_nodes):
        self.graph = graph
        self.init_infected_nodes = set(init_infected_nodes) # will use these nodes to infect their neighbors
            
    
    # spread SI model
    # return the infected_nodes list, in the infection order
    def spread(self, maxN=None):
        visited_edges=set()
        infected_nodes=set()
        bound_nodes=[]
        infected_list=[]
        for u in self.init_infected_nodes:
            infected_nodes.add(u)
            infected_list.append(u)
            for v in self.graph.neighbors(u):
                if v not in self.init_infected_nodes:
                    bound_nodes.append(v)
        if (maxN is None):
            maxN = self.graph.number_of_nodes()
        while (len(infected_list) < maxN):
            # select u from bound_nodes
            idx = np.random.randint(0, len(bound_nodes))
            u = bound_nodes[idx]
            infected_list.append(u)
            infected_nodes.add(u)
            bound_nodes = list(filter(lambda a: a != u, bound_nodes))
            for v in self.graph.neighbors(u):
                if v not in infected_nodes:
                    bound_nodes.append(v)
        return infected_list

def test():
    graph = nx.balanced_tree(2,3)
    nx.draw_networkx(graph)
    plt.show()
    
    simodel2 = SIModel2(graph, [4])
    print(simodel2.spread())
        
test()
```

```python
class RumorCentralityEstimator:
    def __init__(self):
        self.rc_dict = {}
        self.t_dict = {}
        self.p_dict = {}
    
    # return a dict: node_id -> rumor centrality value
    def calcRC(self, graph, infected_list):
        infected_graph = graph.subgraph(infected_list)
        vroot = infected_list[0] # any node as root
        bfs_tree = nx.bfs_tree(infected_graph, source = vroot)
        self.message_pass_algorithm(bfs_tree, vroot)
        return self.rc_dict
    
    def calcRC_BFSheuristic(self, graph, infected_list):
        infected_graph = graph.subgraph(infected_list)
        vroot = infected_list[0] # any node as root
        bfs_tree = nx.bfs_tree(infected_graph, source = vroot)
        self.message_pass_algorithm(bfs_tree, vroot)
        ## BFS heursitic
        for v in infected_list:
            nodes_bfs=[v]
            for edge in nx.bfs_edges(infected_graph, v):
                nodes_bfs.append(edge[1])
            dsum = bfs_tree.degree(nodes_bfs[0])
            if (dsum != 0):
                psig_log = np.log(1/dsum)
            else:
                psig_log = 0
            for i in range(1, len(nodes_bfs)-1):
                dsum += bfs_tree.degree(nodes_bfs[i]) - 2
                if (dsum!=0):
                    psig_log += np.log(1/dsum)
            ## offset  rc_dict by psig_log
            self.rc_dict[v] += psig_log            
        return self.rc_dict
        
    # tree is a nx.Digraph, vroot is its root
    def message_pass_algorithm(self, tree, vroot):
        self.dfs1(tree, vroot)
        self.dfs2(tree, vroot)
    
    
    # dfs from node v to traverse the tree, to fill in t_dict, p_dict (bottom up)
    # note, in p_dict, it actuallys saves loge(p)
    def dfs1(self, tree, v):
        if ( tree.successors(v) == []): # v is leaf 
            self.t_dict[v] = 1
            self.p_dict[v] = np.log(1)
            return
        sum_t = 0
        sum_p = 0
        for u in tree.successors(v):
            self.dfs1(tree, u)
            sum_t += self.t_dict[u]
            sum_p += self.p_dict[u]
        self.t_dict[v] = 1+sum_t # t_dict[v]: number of nodes in the subtree, rooted at v
        self.p_dict[v] = np.log(self.t_dict[v]) + sum_p # p_dict[v] = log(..)
    
    def dfs2(self, tree, v):
        prede = list(tree.predecessors(v))
        if (prede == []): # v is root of tree
            sum_logx = sum([np.log(x) for x in range(1,self.t_dict[v]+1)])
            self.rc_dict[v] = sum_logx - self.p_dict[v]
        else:
            parent = prede[0]
            n=len(self.t_dict)
            self.rc_dict[v] = self.rc_dict[parent] + (np.log((self.t_dict[v]) / (n-self.t_dict[v])))
        for u in tree.successors(v):
            self.dfs2(tree, u)

# test1. 1-2-3. passed
def test1():
    graph = nx.Graph()
    graph.add_nodes_from([1,2,3])
    graph.add_edges_from([[1,2], [2,3]])
    infected = [1,2,3]
    rce = RumorCentralityEstimator()
    rc_dict = rce.calcRC(graph, infected)
    #print(rc_dict)
    print({k:np.exp(rc_dict[k]) for k in rc_dict.keys()})
    
test1()

# test2. passed
# 1 - 2 - 4
#  \3  \5
def test2():
    g = nx.Graph()
    g.add_nodes_from([1, 2, 3, 4, 5])
    g.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5)])
    infected = [1,2,3,4,5]
    rce = RumorCentralityEstimator()
    rc_dict = rce.calcRC(g, infected)
    #print(rc_dict)
    print({k:np.exp(rc_dict[k]) for k in rc_dict.keys()})

test2()

#    1 
# 2    3
#4 5  6 7 
def test3():
    g = nx.Graph()
    g.add_nodes_from([1,2,3,4,5,6,7])
    g.add_edges_from([(1,2), (1,3), (2,4), (2,5), (3,6), (3,7)])
    infected = [1,2,3,4,5,6,7]
    np.random.shuffle(infected)
    rce = RumorCentralityEstimator()
    rc_dict = rce.calcRC(g, infected)
    #print(rc_dict)
    print({k:np.exp(rc_dict[k]) for k in rc_dict.keys()})
test3()
```

```python
# give a graph, do experiment to test performance of RCE (also other centralities for comparison)
# centralities: "RCE", "degree", "harmonic", "between"
# graph: nx.Graph, where SI simulation & RCE is performed
# ntry: how many SI simulation & RCE is performed
# init_node: which node is the source in simulation, None to use random init
# spread_stopN: the N infected nodes to stop
# p_trans: what's the transmission probability
# return correct_counts: metrics -> [counts for different]
def RCEexpt2_varyingN(graph, ntry=100, init_node=None, max_N=None, init_node_random_list=None,
                     metrics=["rumor_centrality", "degree_centrality", "harmonic_centrality"]):
    Nsegs = 20 # divide the |graph| by Nsegs, so stopN = [0.1,0.2,..1.0] * |graph|
    correct_counts = {k:np.zeros((Nsegs,),np.float) for k in metrics}
    if (max_N is None):
        max_N=graph.number_of_nodes()
    metrics=set(metrics)
    for itry in range(ntry):
        print("itry=%d", itry)
        if (init_node is None):
            if (init_node_random_list is None): # random from all ndoes
                init_node = list(graph.nodes())[np.random.randint(0, graph.number_of_nodes())]
            else: # random from init_node_random_list
                init_node_idx = np.random.randint(0, len(init_node_random_list))
                init_node = init_node_random_list[init_node_idx]

        simodel = SIModel2(graph, [init_node])
        # run for all 
        infected_list = simodel.spread(max_N)
        for iseg in range(1,Nsegs+1): 
            segN = int(iseg/Nsegs * max_N)
            if (segN > max_N):
                segN = max_N
            infected_list_top = infected_list[0:segN]
            np.random.shuffle(infected_list_top)
            if ("rumor_centrality" in metrics):
                rce = RumorCentralityEstimator()
                dct_rumor_cen = rce.calcRC(graph, infected_list_top)
                top_node_rc = max([rc for rc in dct_rumor_cen.items()], key=lambda x:x[1])
                if (top_node_rc[0] == init_node):
                    correct_counts["rumor_centrality"][iseg-1] += 1
            if ("rumor_centrality_bfs" in metrics):
                rcex = RumorCentralityEstimator()
                dct_rumor_cen = rcex.calcRC_BFSheuristic(graph, infected_list_top)
                top_node_rc = max([rc for rc in dct_rumor_cen.items()], key=lambda x:x[1])
                if (top_node_rc[0] == init_node):
                    correct_counts["rumor_centrality_bfs"][iseg-1] += 1
            if ("degree_centrality" in metrics):
                # reference -- degree
                cen_degree = nx.degree_centrality(graph.subgraph(infected_list_top))
                top_node_deg = max([cen for cen in cen_degree.items()], key=lambda x:x[1])
                if (top_node_deg[0] == init_node):
                    correct_counts["degree_centrality"][iseg-1] += 1
            if ("harmonic_centrality" in  metrics):
                # reference -- harmonic
                cen_harmonic = nx.harmonic_centrality(graph.subgraph(infected_list_top))
                top_deg_harmonic = max([cen for cen in cen_harmonic.items()], key=lambda x:[1])
                if (top_deg_harmonic[0] == init_node):
                    correct_counts["harmonic_centrality"][iseg-1] += 1  
    print("end")
    return correct_counts
```

```python
# regular tree generator
#level=1, 1 node
# return (graph, dict_height:node->height)
def genRegularTree(k, level):
    graph = nx.Graph()
    dict_height = {}
    iroot, nexti, edges = genEdgeList(0, level, k, dict_height, True)
    graph.add_nodes_from(range(nexti))
    graph.add_edges_from(edges)
    return graph, dict_height

# input, dict_height: node -> height (leaf = 0);
# ouput, rev_dict_hgt: height -> [nodes]
def reverseHeightDict(dict_height):
    rev_dict_hgt = dict()
    for k,v in dict_height.items():
        if (v not in rev_dict_hgt):
            rev_dict_hgt[v] = []
        rev_dict_hgt[v].append(k)
    return rev_dict_hgt

def getInitNodeList(rev_dict_hgt, floor_level):
    res=[]
    for k in rev_dict_hgt.keys():
        if (k>=floor_level):
            res.extend(rev_dict_hgt[k])
    return res

# return root_idx, next_idx, [edge list]
def genEdgeList(starti, level, k, dict_height, isroot=False):
    if (level == 1):
        dict_height[starti] = 0
        return (starti, starti+1, [])
    else:
        idx = starti+1
        edges = []
        dict_height[starti] = level - 1
        if (isroot):
            m=k
        else:
            m=k-1
        for i in range(m):
            ichild, inext, edges1 = genEdgeList(idx, level-1, k, dict_height)
            edges.extend(edges1)
            edges.append([starti, ichild])
            idx = inext
        return (starti, idx, edges)

#test
tree3, dict_hgt = genRegularTree(3,3)
print(tree3.number_of_nodes())
print(dict_hgt)
print(reverseHeightDict(dict_hgt))
nx.draw_networkx(tree3)
plt.show()
nx.degree(tree3)
```

```python
# try regtree
level=18
k=3
regtree, dict_hgt = genRegularTree(k, level)
rev_dict_hgt = reverseHeightDict(dict_hgt)
# use nodes with level >= 14 as init_node
init_node_list = getInitNodeList(rev_dict_hgt, 3) ## select tiers>=3, which is level-3 = 15 (top 15 tiers)
print(len(init_node_list))
# 
max_N=500
n_try=20
correct_counts = RCEexpt2_varyingN(regtree, ntry=n_try, max_N=max_N, 
                init_node_random_list=init_node_list, metrics=["rumor_centrality", "harmonic_centrality"])

print(regtree.number_of_nodes())
# plot curve (x=N ratio, y=correct ratio)
#plt.figure(figsize=(8,6))
xs = max_N * np.arange(0.05, 1.01,0.05)
symbol={"rumor_centrality": "*", "harmonic_centrality": "*"}
for key in correct_counts.keys():
    plt.plot(xs, correct_counts[key]/n_try, symbol[key], label=key)
plt.legend(fontsize=12)
#plt.ylim([-0.04,0.4])
plt.title("Regular tree (d=3), start from inner 15 tiers", fontsize=14)
plt.xlabel("N. number of infected nodes", fontsize=14)
plt.ylabel("Accuracy", fontsize=14)
plt.show()
```

```python
# SBM graph
# e in [0,2c], n must be even
def sbm_partile(n, c, e):
    p_in = c/n + e/(2*n)
    p_out = c/n - e/(2*n)
    W=[[p_in, p_out], [p_out, p_in]]
    G = nx.stochastic_block_model([int(n/2),int(n/2)], W)
    return G

c=10
n=1000
ans={}
for e in [0, 5, 10, 15, 19]:
    sbm2 = sbm_partile(n, c, e)
    n_try=100
    max_N=200
    correct_counts = RCEexpt2_varyingN(sbm2, ntry=n_try, max_N=max_N, 
                init_node_random_list=None, metrics=["rumor_centrality"])
    ans[e] = correct_counts["rumor_centrality"]

    # plot curve (x=N ratio, y=correct ratio)
#plt.figure(figsize=(8,6))
xs = max_N * np.arange(0.05, 1.01,0.05)
ref_ys = 1/xs
plt.plot(xs, ref_ys, '--k', label="1/N")
symbol=defaultdict(lambda: "*-")
for key in ans.keys():
    if (key>=0):
        plt.plot(xs, ans[key]/n_try, symbol[key], label="e=%d" %key)
plt.legend(fontsize=12)
#plt.ylim([-0.04,0.4])
plt.title("SBM, n=200, c=10", fontsize=14)
plt.xlabel("N. number of infected nodes", fontsize=14)
plt.ylabel("Accuracy", fontsize=14)
plt.show()
```
